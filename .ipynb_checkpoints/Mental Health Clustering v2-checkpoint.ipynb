# Mental Health Clustering

### Data Description

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from prince import MCA
from sklearn.cluster import KMeans

import joblib
```

```python
data = pd.read_csv("Mental Health Dataset.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

**Inference**

* Dataset consists of categorical self-reported mental health indicators
* No numeric severity â†’ must be engineered
* Suitable for MCA + clustering with guardrails

### Data Description & Column Selection

```python
data.isnull().sum()
```

### Missing Value Handling

```python
data = data.fillna("Unknown")
```

**Inference:** MCA and KMeans cannot handle NaN, so missing values are treated as a valid category.

### Severity Mapping

```python
severity_map = {
    # Binary
    "Yes": 1,
    "No": 0,

    # Stress
    "Manageable": 0,
    "Elevated": 1,
    "Overwhelming": 2,

    # Habit Change
    "No noticeable changes": 0,
    "Some changes": 1,
    "Significant changes": 2,

    # Mood
    "Rarely": 0,
    "Sometimes": 1,
    "Often": 2,

    # Coping
    "Coping well": 0,
    "Struggling at times": 1,
    "Struggling most of the time": 2,

    # Work
    "Highly engaged": 0,
    "Somewhat engaged": 1,
    "Not engaged": 2,

    # Social
    "As connected as usual": 0,
    "Slightly less connected": 1,
    "Much less connected": 2,

    # Fallback
    "Unknown": 1
}
```

**Inference:** Maps qualitative responses to severity-aware numeric scores.

### Compute Mental Distress Index

```python
core_symptoms = [
    "Growing_Stress",
    "Changes_Habits",
    "Mood_Swings",
    "Coping_Struggles",
    "Work_Interest",
    "Social_Weakness"
]

data["MDI"] = data[core_symptoms].apply(
    lambda row: sum(severity_map.get(value, 1) for value in row),
    axis=1
)

data[["MDI"]].head()
```

### Visualize MDI Distribution

```python
# KDE Plot

plt.figure(figsize=(8,4))
sns.kdeplot(data["MDI"], fill=True)
plt.title("MDI Distribution")
plt.xlabel("MDI Score")
plt.show()
```

```python
# Boxplot (Outlier Understanding)

plt.figure(figsize=(6,4))
sns.boxplot(x=data["MDI"])
plt.title("MDI Boxplot")
plt.show()
```

### Assign Risk Bands

```python
def assign_risk_band(mdi):
    if mdi >= 8:
        return "High"
    elif mdi >= 4:
        return "Moderate"
    else:
        return "Low"

data["risk_band"] = data["MDI"].apply(assign_risk_band)
data["risk_band"].value_counts()
```

**Inference:** Risk banding ensures ethical interpretation before clustering.

### MCA Transformation

```python
mca = MCA(
    n_components=5,
    random_state=42
)

X_mca = mca.fit_transform(data[core_symptoms])
X_mca.head()
```

**Inference:** MCA preserves relationships in categorical mental-health data better than PCA.

### Risk-Aware KMeans Clustering

```python
from sklearn.cluster import KMeans

cluster_models = {}

for band in ["Low", "Moderate", "High"]:
    band_mask = data["risk_band"] == band
    sample_count = band_mask.sum()

    print(f"{band} risk samples: {sample_count}")

    if sample_count == 0:
        print(f"âŒ No samples for {band}, skipping")
        continue

    # ðŸ‘‰ Adaptive clusters
    n_clusters = 2 if sample_count >= 10 else 1

    X_band = X_mca.loc[band_mask].copy()
    X_band.iloc[:, :3] *= 2  # weight core symptom components

    kmeans = KMeans(
        n_clusters=n_clusters,
        random_state=42
    )
    kmeans.fit(X_band)

    cluster_models[band] = kmeans
```

**Inference:** Clusters are trained within each risk band, not across the entire population.

```python
# Step 1: Create a GLOBAL fallback model if High-risk is missing
if "High" not in cluster_models:
    print("âš ï¸ High-risk samples missing â†’ training GLOBAL fallback model")

    X_global = X_mca.copy()
    X_global.iloc[:, :3] *= 2

    fallback_kmeans = KMeans(
        n_clusters=2 if len(X_global) >= 10 else 1,
        random_state=42
    )
    fallback_kmeans.fit(X_global)

    cluster_models["High"] = fallback_kmeans

# Step 2: Ensure Moderate exists
if "Moderate" not in cluster_models:
    print("âš ï¸ Moderate model missing â†’ using High-risk model")
    cluster_models["Moderate"] = cluster_models["High"]

# Step 3: Ensure Low exists
if "Low" not in cluster_models:
    print("âš ï¸ Low model missing â†’ using High-risk model")
    cluster_models["Low"] = cluster_models["High"]

print("âœ… Final cluster models:", cluster_models.keys())
```

## Deployment

```python
joblib.dump(mca, "mca_transformer.joblib")
joblib.dump(cluster_models, "risk_band_cluster_models.joblib")
joblib.dump(severity_map, "severity_map.joblib")
```
